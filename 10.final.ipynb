## 墨尔本每日最低温度 
### 问题 

(1) 剔除国庆节、五一劳动节和每月第一个周一，求每月的平均最低气温。 

(2) 季节指数是一种对于周期性变化序列的特征刻画。记数据集中第 k 年 平均最低气温为 TYk(k = 1, ..., 10)，第 k 年第 j 个月平均最低气温为 。请按照如上定义，计算 12 个 ∑ TMkj TM (j = 1, ..., 12)，定义 S = k kj j 月的季节指数 Sj。

(3) 移动平均法是一种时间序列的常见平滑方式，可分为 k 期移动平均和 k 期中心移动平均，都使用了某一时刻及其周围的数据对该时刻的数据进行平滑修正。

```python
#读取数据表
#读取数据表
df_melbourne = pd.read_csv('./data/墨尔本温度数据.csv')#墨尔本1981-1990共十年最低气温数据
df_melbourne.head()
```

```python
df_melbourne['Date'] = pd.to_datetime(df_melbourne['Date']) #转换为时间数据格式to_datetime
df_melbourne = df_melbourne.set_index('Date',drop=True)
df_melbourne.index
```

```python
# (1) 剔除国庆节、五一劳动节和每月第一个周一，求每月的平均最低气温。

# 1999年9月,国务院改革出台新的法定休假制度,每年国庆节、春节和“五一”法定节日加上倒休,全国放假7天。
# 假设本题中国庆节、五一劳动节特指10.1、5.1两天

d = pd.date_range(start='1981-1-1',end='1990-12-31')
mask1 = (d.weekday == 0) & (0< d.day) & (d.day < 8)
mask10 = (d.month==10)&(d.day==1) 
mask5 = (d.month==5)&(d.day==1) 
holidays = d[mask1].append(d[mask5]).append(d[mask10]).sort_values()
weekmask = 'Sun Mon Tue Wed Thu Fri Sat'

# bdate_range是一个类似与date_range的方法，特点在于可以在自带的工作日间隔设置上，再选择weekmask参数和holidays参数 
# 它的freq中有一个特殊的'C'/'CBM'/'CBMS'选项，表示定制，需要联合weekmask参数和holidays参数使用
melbourne = df_melbourne.copy().reindex(pd.bdate_range(start='1981-1-1',end='1990-12-31',freq='C',weekmask = weekmask,holidays=holidays))
#C是 CDay(定制工作日)
#求每月的平均最低气温
melbourne.resample('M').mean().head()   
#以月份resample('M')重采样。
# 所谓重采样，就是指resample函数，它可以看做时序版本的groupby函数
```

```python
#(2) 季节指数,月均值/年均值
TY   = df_melbourne.resample('Y').mean().sum()
TMkj = df_melbourne.resample('M').mean()

for i in range(12):
    j = i+1
    print('S{}='.format(j),(TMkj[TMkj.index.month==j].sum()/TY).values[0].round(decimals=4))
```

```python
#(3) 移动平均
#(a) 求原序列的 5 期移动平均序列(自身与前四个的均值)
df_melbourne.rolling(window=5,min_periods=-5).sum()/5
#所谓rolling方法，就是规定一个窗口，它和groupby对象一样，本身不会进行操作，需要配合聚合函数才能计算结果
#min_periods参数是指需要的非缺失数据点数量阀值  
```

```python
#(b) 原序列的 5 期中心移动平均序列。
df_melbourne.rolling(window=5,min_periods=-5).sum().shift(-2)/5
#原序列的 6 期中心移动平均序列。
(df_melbourne.shift(-3)/2+df_melbourne.rolling(window=5,min_periods=-5).sum().shift(-2)+df_melbourne.shift(3)/2)/6
```

## 2016 年 8 月上海市摩拜单车骑行记录
### 问题

(1) 平均而言，周末单天用车量比工作日单天用车量更大吗?

(2) 工作日每天的高峰时间段大致为上午 7:30 至 9:30、下午 17:00 至 19:00， 请问 8 月里早高峰骑行记录量(以 start_time 为准)高于晚高峰的有 几天?

(3) 请给出在所有周五中(以 start_time 为准)，记录条数最多的那个周五 所在的日期，并在该天内分别按 30 分钟、2 小时、6 小时统计摩拜单车 使用时间的均值。

(4) 请自行搜索相关代码或调用库，计算每条记录起点到终点的球面距离。

(5) 摩拜单车的骑行结束时间是以电子锁关闭的记录时间为准，但有时候用 户会忘记关锁，导致骑行时间出现异常。同时，正常人的骑行速度往往 大致落在一个合理的区间，请结合上一问中的球面距离和骑行起始、结 束时间，找出潜在的异常骑行记录。

(6) 由于路线的曲折性，起点到终点的球面距离往往不能充分反应行程长 度，请利用 track 列的路线坐标数据，计算估计实际骑行距离，并重新 仿照上一问的方法找出可能的异常记录。

```python
df_mobike.start_time = pd.to_datetime(df_mobike.start_time) 
df_mobike.end_time = pd.to_datetime(df_mobike.end_time) 
df_mobike = df_mobike.set_index('start_time',drop=False)  #这一步设置了index为'start_time'，所以下面的.index获取的是开始时间

# DataFrame可以通过set_index方法，设置单索引和复合索引。 
# DataFrame.set_index(keys, drop=True, append=False, inplace=False, verify_integrity=False) 
# reset_index可以还原索引，重新变为默认的整型索引 
# DataFrame.reset_index(level=None, drop=False, inplace=False, col_level=0, col_fill=”) 
# level控制了具体要还原的那个等级的索引 
# drop为False则索引列会被还原为普通列，否则会丢失
```

```python
#（1) 平均而言，周末单天用车量比工作日单天用车量更大吗?
#freq参数 D/B：日/工作日 通过加入周末与只计算工作日的平均值的大小来对比？
print('周末单天用车量比工作日单天用车量更大吗?',df_mobike.resample('D').count().mean()[0]>df_mobike.resample('B').count().mean()[0])
```

```python
#(2) 工作日每天的高峰时间段大致为上午 7:30 至 9:30、下午 17:00 至 19:00， 
# 请问 8 月里早高峰骑行记录量(以 start_time 为准)高于晚高峰的有 几天?

highday=[]
for i in range(31):
    day = '2016-8-{}'.format(i+1) #‘{}’.format(i+1)表示的是i+1替代方括号被写入
    if(pd.to_datetime(day).dayofweek>4):continue#跳过周末
    morning   = df_mobike[(df_mobike.index>='{} 07:30'.format(day))&( df_mobike.index<='{} 09:30'.format(day))].count()[0]
    #df_mobike.index就是时间？[59]那步改了
    afternoon = df_mobike[(df_mobike.index>='{} 17:00'.format(day))&( df_mobike.index<='{} 19:00'.format(day))].count()[0]
    if(morning>afternoon):
        highday.append(day)
print("8 月早高峰骑行记录量高于晚高峰的有{}天,分别是{}".format(len(highday),highday))
```

```python
#(3) 请给出在所有周五中(以 start_time 为准)，记录条数最多的那个周五 所在的日期，
friday = df_mobike[df_mobike.index.dayofweek==4]['orderid'].resample('D').count().idxmax()
print('条数最多的那个周五是',friday.date())

df_mobike_friday = df_mobike[str(friday.date())].copy()
df_mobike_friday['usetime'] = (df_mobike_friday.end_time-df_mobike_friday.start_time).dt.total_seconds()
#采用dt对象可以轻松获得关于时间的信息，转为绝对秒数求均值
#并在该天内分别按 30 分钟、2 小时、6 小时统计摩拜单车 使用时间的均值。

#Timedelta绝对时间差的特点指无论是冬令时还是夏令时，增减1day都只计算24小时 
#分别以30min，2小时，6小时对使用时间进行采样，每一个都采样五个点
display( 
    df_mobike_friday['usetime'].resample('30min').mean().apply(lambda x:pd.Timedelta(seconds = int(x))).head(),
    df_mobike_friday['usetime'].resample('2h'   ).mean().apply(lambda x:pd.Timedelta(seconds = int(x))).head(),
    df_mobike_friday['usetime'].resample('6h'   ).mean().apply(lambda x:pd.Timedelta(seconds = int(x))).head(),
)   
```

```python
#（4）起点终点球面距离
from geopy.distance import geodesic 
df_mobike_dist = df_mobike.copy() 
df_mobike_dist.head()
```

```python
#计算起始点距离，单位m米
df_mobike_dist['se_dist'] = df_mobike_dist.apply(lambda x: geodesic((x.start_location_y,x.start_location_x),(x.end_location_y,x.end_location_x)).m,axis=1)
```

```python
df_mobike_dist.head()
```

```python
#(5)请结合上一问中的球面距离和骑行起始、结束时间，找出潜在的异常骑行记录。
df_mobike_dist['usetime'] = (df_mobike_dist.end_time-df_mobike_dist.start_time).dt.total_seconds()#转为绝对秒数求均值
df_mobike_dist['se_speed'] = df_mobike_dist.apply(lambda x:x.se_dist/x.usetime,axis=1)
#x.se_dist/x.usetime 直接利用整列相除
```

```python
print('mean',df_mobike_dist.se_speed.mean())
for q in [0.95,0.75,0.5,0.25,0.1,0.05,0.04,0.03,0.02,0.01,0.001]:
    print(q*100,'分位数',df_mobike_dist.se_speed.quantile(q)) 
# xx分位数就是速度乘以从[0.95,0.75,0.5,0.25,0.1,0.05,0.04,0.03,0.02,0.01,0.001]中选出 xx？
    
#可能忘记锁车的：速度很小很小，代表后面一直没动，所以忘记锁车了
df_mobike_dist[df_mobike_dist.se_speed<df_mobike_dist.se_speed.quantile(0.05)].head()
```

```python
#(6) 利用 track 列的路线坐标数据，计算估计实际骑行距离，并重新仿照上一问的方法找出可能的异常记录。
track = '121.367,31.238#121.367,31.239#121.368,31.240#121.368,31.241#121.368,31.242#121.368,31.252#121.369,31.242#121.369,31.243#121.369,31.244#121.369,31.245#121.369,31.252#121.370,31.245#121.370,31.246#121.370,31.247#121.370,31.251#121.370,31.252#121.371,31.248#121.371,31.249#121.371,31.250#121.371,31.251'
def getdist(track):
    tlist=[]
    dist = 0
    for t in track.split('#'):
        for tt in t.split(','):
            (tlist.append(tt))
    for i in range(int(len(tlist)/2)-1):
        try:
            dist+=geodesic((tlist[i*2+1],tlist[i*2]),(tlist[i*2+3],tlist[i*2+2])).m
        except ValueError:
            dist+=0 
    return dist 
getdist(track)
```

```python
df_mobike_dist['track_dist']=df_mobike_dist.track.apply(lambda x:getdist(x))
df_mobike_dist['track_speed'] = df_mobike_dist.apply(lambda x:x.track_dist/x.usetime,axis=1)
print('mean',df_mobike_dist.track_speed.mean())
for q in [0.95,0.75,0.5,0.25,0.1,0.05,0.04,0.03,0.02,0.01,0.001]: 
    print(q*100,'分位数',df_mobike_dist.track_speed.quantile(q))  
```
